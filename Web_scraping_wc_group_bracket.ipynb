```python
# Importing packages

import numpy as np
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup as Soup
```

```python
# Iniciating variables
column = ['Nation', 'Position', 'Player', 'Club',
         'Age', 'Caps']
wc_squads = pd.DataFrame(columns = column)

# Web scraping squads from WC 2022
url = "https://en.wikipedia.org/wiki/2022_FIFA_World_Cup_squads"
p_html = requests.get(url)
p_soup = p_html.text
data = Soup(p_soup,'html.parser')
cont = 0

for i in data.findAll('tbody'):
    h3 = data.findAll('h3')
    
    for j in i.findAll('tr'):
        td = j.findAll('td')
        th = j.find('th')
        if(len(td)) == 0:
            continue
        Nation = h3[cont].text.split("[")[0]
        Position = td[1].find('a').get('title').split()[0]
        Player = th.find('a').text
        Club = td[5].find_all('a')[1].text
        Age = td[2].text.split()[-1][0:2]
        Caps = td[3].text.split("\n")[0]
        
        player_data = pd.DataFrame([[Nation, Position, Player, Club,
         Age, Caps]])
        player_data.columns = column
        wc_squads = wc_squads.append(player_data, ignore_index = True)
        
    cont = cont + 1
```

```python
# Saving csv file
wc_squads.to_csv('wc_squads_2022.csv', index = None)
```
